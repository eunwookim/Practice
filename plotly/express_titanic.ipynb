```python
import seaborn as sns
import pandas as pd
import numpy as np
import plotly.express as px

tit = sns.load_dataset('titanic')
```

```python
tit.info()
```

```python
pclass_fare = tit[['pclass','fare']]
```

```python
pclass_fare
```

```python
a = px.bar(pclass_fare,x='pclass',y='fare')
a
```

```python
pclass_survived = tit[['pclass','survived']]
print(pclass_survived.info())
pclass_survived
```

```python
b= px.bar(pclass_survived,x='pclass',y='survived')
b
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

x=make_subplots(rows=1,cols=2,shared_yaxes=True)
Fare = go.Scatter(x=pclass_fare['pclass'],y=pclass_fare['fare'],mode='markers')
Survived = go.Scatter(x=pclass_survived['pclass'],y=pclass_survived['survived'],mode='markers')
x.add_trace(Fare, row=1,col=1)
x.add_trace(Survived,row=1,col=2)
# 수치들이 다양하지 않으면 산점도 그래프를 쓰는 것이 유용하지 않다.
```
